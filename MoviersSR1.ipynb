```python
movies = movies[['Movie_Title','Year','Actors','Rating','main_genre','side_genre']]
```

```python
#movies['genres'] = movies['main_genre']+ ','+ movies['side_genre']

movies['genres'] = movies['main_genre'].str.cat(movies['side_genre'],sep = ',')
```

```python
movies.head()
```

```python
del movies['main_genre']
del movies['side_genre']
```

```python
movies.head()
```

```python
movies = movies[['Movie_Title','Year','Actors','Rating','genres']]
```

```python
movies.head()
```

```python
movies['genres']
```

```python
import ast
from ast import literal_eval
```

```python
movies['genres'] = movies['genres'].str.strip('()').str.split(',')
```

```python
movies.head()
```

```python
movies.head()
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['genres'] = movies['genres'].apply(collapse)
movies['Actors'] = movies['Actors'].apply(collapse)
```

```python
movies.head()
```

```python
movies['tags'] = movies['Actors'] + movies['genres'] 
movies.head()
```

```python
movies = movies[['Movie_Title','Actors','genres','tags']]
movies.head()
```

```python
new_df = movies[['Movie_Title','tags']]
new_df
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5600,stop_words='english')
```

```python
vector = cv.fit_transform(new_df['tags']).toarray()
```

```python
vector[0]
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []

    for i in text.split():
       y.append(ps.stem(i))

    return " ".join(y)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity[2]
```

```python
def recommend(movie):
    index = new_df[new_df['Movie_Title'] == movie].index[0]
    distances = similarity[index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key = lambda x:x[1])[1:10]
                         
    for i in movies_list:
        print(new_df.iloc[i[0]].Movie_Title)
```

```python
recommend('The Dark Knight')
```
